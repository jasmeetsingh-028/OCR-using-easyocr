```python
import easyocr #for optical character recognition
import cv2  #for loading and processing on image files
import numpy as np
from matplotlib import pyplot as plt #plot
import math
import os
```

# Importing images and performing OCR

```python
img_dir='C:\\Users\\sjasm\\Images\\ocr images'  #IMAGE DIRECTORY
results=[]
image_path=[]


for files in os.listdir(img_dir):
    img_path=os.path.join(img_dir,files)
    image_path.append(img_path)
    reader = easyocr.Reader(['en'], gpu=False)
    results.append(reader.readtext(img_path))    
```

```python
#results
```

```python
coords=[]
for result in results:
    for coord in result:
        coords.append(coord[0])
```

```python
dims=[]
for coord in coords:
    left1=int(coord[0][0])
    left2=int(coord[0][1])
    right1=int(coord[2][0])
    right2=int(coord[2][1])
    arr=[(left1,left2),(right1,right2)]
    dims.append(arr)
```

## Results

```python
i=0
j=0
color = (0, 255, 0)
for result in results:
    print("Image {}:".format(j+1))
    for text in result:
        path=image_path[j]
        image = cv2.imread(path)
        image = cv2.rectangle(image,dims[i][0] , dims[i][1], color, 5)
        plt.imshow(image)
        plt.show()
        print("text found: ", text[1])
        print("top left coordinate: ",dims[i][0],"   bottom right coordinate: ",dims[i][1])
        i+=1
    j+=1
```
